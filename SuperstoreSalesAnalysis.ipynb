```python
#null olan posta kodlarını listele
df[df["Postal Code"].isnull()]
```

```python
#null olan alanları 5402 ile doldur
df["Postal Code"] = df["Postal Code"].fillna(5402)
```

```python
df
```

```python
#her bir sütundaki null olan hücre sayısı
df.isnull().sum()
```

```python
#verileri doğru bir şekilde girmiş miyiz diye kontrolü sağlıyoruz
df[df["City"]=="Burlington"]
```

```python
#ay sütunu
df["Month"]=df["Order Date"].str[3:5]
df
```

```python
# Ayların başındaki sıfırları ortadan kaldırmak ve tam sayılara dönüştürmek
df["Month"] = df["Month"].astype("int32")
print(df["Month"].dtype)
df
```

```python
df["Month"].dtype
```

```python
# Dictionary to map month numbers to month names in Turkish
d = {
    1: 'ocak', 2: 'şubat', 3: 'mart', 4: 'nisan', 5: 'mayıs', 6: 'haziran',
    7: 'temmuz', 8: 'ağustos', 9: 'eylül', 10: 'ekim', 11: 'kasım', 12: 'aralık'
}
```

```python
d[4]
```

```python
# Converting month numbers to month names
df["Month"] = [d[i] for i in df["Month"]]
```

```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
df
```

```python
df["Ship Date"]=pd.to_datetime(df["Ship Date"])
df
```

```python
df["Year"] = df["Order Date"].dt.year
df
```

```python
#hangi yıldan kaç tane veri var diye bakıyoruz
df["Year"].value_counts()
```

```python
#hangi ayda kaç tane veri var diye bakıyoruz(alışveris sayısı)
df["Month"].value_counts()
```

```python
#en fazla satış yapılan ay
months_df = df.groupby("Month").sum()
months_df
```

```python
#aylara göre toplam satış miktarları
months_df = months_df[["Sales"]]
months_df
```

```python
months_df.reset_index(inplace = True)
months_df
```

```python
plt.bar(months_df["Month"],months_df["Sales"])
plt.xlabel("Aylar")
plt.ylabel("satis miktarı")
plt.show()
```

```python
plt.bar(months_df["Month"],months_df["Sales"])
plt.xlabel("Aylar")
plt.ylabel("satis miktarı")
plt.xticks(fontsize=14, rotation=90)
plt.yticks(fontsize=14)
plt.show()
```

```python
#en fazla satış yapılan yıl
year_df = df.groupby("Year").sum()
year_df
```

```python
#aylara göre toplam satış miktarları
year_df = year_df[["Sales"]]
year_df
```

```python
year_df.reset_index(inplace = True)
year_df
```

```python
plt.bar(year_df["Year"],year_df["Sales"])
plt.xlabel("Yıllar")
plt.ylabel("satis miktarı")
plt.xticks(fontsize=14, rotation=90)
plt.yticks(fontsize=14)
plt.show()
```

```python
#en cok para harcayan müsteri kimdir#en fazla satış yapılan yıl
costumer_df = df.groupby("Customer Name").sum().sort_values("Sales",ascending=False)
#kullanmayacağımız alanlar kaldırıyoruz
costumer_df.drop(["Postal Code","Year"],axis=1,inplace=True)
costumer_df
```

```python
#yanlara index ekleyelim
costumer_df.reset_index(inplace=True)
costumer_df
```

```python
#ilk 20 kişinin satış miktarlarının listelenmesi
plt.bar(costumer_df["Customer Name"],costumer_df["Sales"])
plt.xlabel("Ad Soyad")
plt.ylabel("satis miktarı")
plt.xticks(fontsize=14, rotation=90)
plt.yticks(fontsize=14)
plt.show()
```

```python
category = df.groupby("Category").sum()
category
```

```python
#Kategorilere göre satış miktarlarının listelenmesi
plt.bar(category["Category"],category["Sales"])
plt.xlabel("Kategori")
plt.ylabel("satis miktarı")
plt.xticks(fontsize=14, rotation=90)
plt.yticks(fontsize=14)
plt.show()
```
